```python
# 서울 행정구역 파일
seoul_gdf = gpd.read_file("https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json")
```

```python
seoul_gdf
```

```python
# 좌표 -> Point 객체로
def get_point(row):
  return Point(row['경도'], row['위도'])
```

```python
aug["geometry"] = aug.apply(get_point, axis=1)
```

```python
import numpy as np
```

```python
def create_district_column(df):
  districts = []
  for i in range(df.shape[0]):
    d = seoul_gdf.loc[seoul_gdf.geometry.contains(df["geometry"][i]), 'name'].tolist()
    if len(d) == 1:
      d = d[0]
    elif len(d) == 0:
      d = np.NaN
    districts.append(d)
  df["지역구"] = districts
```

```python
seoul_gdf.loc[seoul_gdf.geometry.contains(aug["geometry"].iloc[0]), 'name']
```

```python
create_district_column(aug)
```

```python
aug
```

```python
# 서울 외에 있는 지하철역은 제거해줌
aug.loc[aug["지역구"].isna()]
```

```python
aug.dropna(subset=["지역구"], inplace=True)
```

```python
aug
```

```python
sept["geometry"] = sept.apply(get_point, axis=1)
oct["geometry"] = oct.apply(get_point, axis=1)
create_district_column(sept)
create_district_column(oct)
sept.dropna(subset=["지역구"], inplace=True)
oct.dropna(subset=["지역구"], inplace=True)
```
