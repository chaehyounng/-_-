```python
def dfcmon(df, month, line):
    df=df.drop(columns=["노선", "기종점", "정류장순번"])
    for i in range(df.shape[0]):
        if df["정류장"].iloc[i] == 0:
            df["정류장"].iloc[i] = df["정류장"].iloc[i-1]
        else:
            continue
    df = df[df["이용자유형"]=="장애인"]

    df["노선"] = line
    df["월"]=month
    df["정류장"] = df["정류장"].apply(lambda x:x[:-6])
    df = df.reset_index(drop=True)

    return(df)
```

```python
aug1 = dfcmon(aug1, 8, 1)
aug2 = dfcmon(aug2, 8, 2)
aug3 = dfcmon(aug3, 8, 3)
aug4 = dfcmon(aug4, 8, 4)
aug5 = dfcmon(aug5, 8, 5)
```

```python
sept1 = dfcmon(sept1, 9, 1)
sept2 = dfcmon(sept2, 9, 2)
sept3 = dfcmon(sept3, 9, 3)
sept4 = dfcmon(sept4, 9, 4)
sept5 = dfcmon(sept5, 9, 5)
```

```python
oct1 = dfcmon(oct1, 10, 1)
oct2 = dfcmon(oct2, 10, 2)
oct3 = dfcmon(oct3, 10, 3)
oct4 = dfcmon(oct4, 10, 4)
oct5 = dfcmon(oct5, 10, 5)
```

```python
aug1_5 = pd.concat([aug1, aug2, aug3, aug4, aug5], ignore_index=True)
```

```python
sept1_5 = pd.concat([sept1, sept2, sept3, sept4, sept5], ignore_index=True)
```

```python
oct1_5 = pd.concat([oct1, oct2, oct3, oct4, oct5], ignore_index=True)
```
