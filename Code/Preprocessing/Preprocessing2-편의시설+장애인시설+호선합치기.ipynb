## 주거시설 데이터


```python
shp = gpd.read_file("TL_DSPSN_FCLTY_ST_2019.shp",encoding='cp949')
shp
# 위도와 경도로 변환 (EPSG:4326은 일반적인 위도와 경도 좌표 체계)
shp = shp.to_crs(epsg=4326)
```

```python
shp
```

```python
shp['위도'] = shp['geometry'].y
shp['경도'] = shp['geometry'].x
```

```python
#편의시설 데이터의 위도경도로만 기준점 딕셔너리 생성하기
pointshp = shp[['BD_07_NM','경도', '위도']].drop_duplicates().to_dict(orient='records')

#pointutil=pd.DataFrame(list(pointutil), columns=['콘텐츠명\n(필수 입력)','좌표[X]\n(조건부 선택 입력)', '좌표[Y]\n(조건부 선택 입력)'])
pointshp
```

```python
for base_point in pointshp:
    base_point_geom = Point(base_point['경도'], base_point['위도'])
    gdfsub[f'distance_to_base_point_{pointshp.index(base_point)}'] = gdfsub.geometry.distance(base_point_geom)
    closest_row = gdfsub.loc[gdfsub[f'distance_to_base_point_{pointshp.index(base_point)}'].idxmin(), '정류장']
    shp.at[pointshp.index(base_point),'closest_row']=closest_row
    
    print("편의시설 " + pointshp[pointshp.index(base_point)]['BD_07_NM'] + "에 가장 가까운 정류장:\n" + closest_row + "\n")
```

```python
#지하철 역마다 가까운 편의시설 개수 구하기
count1= shp['closest_row'].value_counts()
count1.sum()
df_count1 = count1.reset_index()
df_count1.columns = ['closest_row', '주변_장애인시설_개수']
df_count1
count1.sum()
```

```python
oct_sub2=pd.merge(oct_sub1,df_count1,how='outer',left_on='정류장',right_on='closest_row')
oct_sub2=oct_sub2.drop('closest_row', axis=1)
oct_sub2['주변_장애인시설_개수'].fillna(0, inplace=True)
oct_sub2
```

```python
sept_sub2=pd.merge(sept_sub1,df_count1,how='outer',left_on='정류장',right_on='closest_row')
sept_sub2=sept_sub2.drop('closest_row', axis=1)
sept_sub2['주변_장애인시설_개수'].fillna(0, inplace=True)
sept_sub2
```
