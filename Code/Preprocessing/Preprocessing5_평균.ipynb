```python
final = oct[['정류장', '호선_개수', '주변_편의시설_개수', '노선', '위도', '경도', '주변_장애인시설_개수', '지역구']]
```

```python
final["주변_편의시설_개수"] = final["주변_편의시설_개수"].astype(int)
final["주변_장애인시설_개수"] = final["주변_장애인시설_개수"].astype(int)
```

```python
aug_n = aug[["정류장", "합계"]]
aug_n.rename(columns={"합계":"8월_장애인_승객수"}, inplace=True)
sept_n = sept[["정류장", "합계"]]
sept_n.rename(columns={"합계":"9월_장애인_승객수"}, inplace=True)
oct_n = oct[["정류장", "합계"]]
oct_n.rename(columns={"합계":"10월_장애인_승객수"}, inplace=True)
```

```python
final = pd.merge(final, aug_n, on="정류장")
final = pd.merge(final, sept_n, on="정류장")
final = pd.merge(final, oct_n, on="정류장")
```

```python
final
```

```python
final["평균_장애인_승객수"] = (final["8월_장애인_승객수"] + final["9월_장애인_승객수"] + final["10월_장애인_승객수"])/3
```
