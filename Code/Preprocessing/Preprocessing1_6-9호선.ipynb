```python
def dfcmon(df,mon):
  df=df.drop(columns=["노선", "기종점", "정류장순번"])
  for i in range(df.shape[0]):
    if df["정류장"].iloc[i] == 0:
        df["정류장"].iloc[i] = df["정류장"].iloc[i-1]
    else:
        continue
  df = df[df["이용자유형"]=="장애인"]

  df["노선"] = df["정류장"].apply(lambda x:x[-4])
  df["월"]=mon
  df["정류장"] = df["정류장"].apply(lambda x:x[:-6])
  df = df.reset_index(drop=True)

  return(df)

```

```python
oct8 = pd.read_excel("/content/drive/MyDrive/턱수염/10월_8호선.xlsx")
oct8=dfcmon(oct8,10)
oct8.head(5)
```

```python
oct6 = pd.read_excel("/content/drive/MyDrive/턱수염/10월_6호선.xlsx")
oct6=dfcmon(oct6,10)
oct6
```
