## 서울시 장애인 편의시설 수

```python
seoul =  folium.Map(location=[37.5665, 126.9780], zoom_start=12)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul)

marker_cluster=MarkerCluster().add_to(seoul)

for lat, long in zip(facilities["위도"], facilities["경도"]):
  folium.Marker([lat, long], icon=folium.Icon(color="green")).add_to(marker_cluster)

seoul
```

```python
seoul_group_data = facilities.groupby("주소[구]\n(조건부 선택 입력)")["콘텐츠 ID\n(필수 입력)"].count()
seoul_group_data
```

```python
seoul_group =  folium.Map(location=[37.5665, 126.9780], zoom_start=11.5)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul_group)

folium.Choropleth(geo_data = seoul_geo,
                  name = "장애인 편의시설 수",
                  data=seoul_group_data,
                  fill_color="YlOrRd",
                  fill_opacity=0.5,
                  line_opacity=0.2,
                  key_on="properties.name",
                  legend_name="장애인 편의시설 수").add_to(seoul_group)

seoul_group
```

```python
bins = list(seoul_group_data.quantile([0, 0.25, 0.5, 0.75, 1]))

seoul_group =  folium.Map(location=[37.5665, 126.9780], zoom_start=11.5)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul_group)

folium.Choropleth(geo_data = seoul_geo,
                  name = "장애인 편의시설 수",
                  data=seoul_group_data,
                  fill_color="YlOrRd",
                  fill_opacity=0.5,
                  line_opacity=0.2,
                  key_on="properties.name",
                  legend_name="장애인 편의시설 수",
                  bins=bins).add_to(seoul_group)

seoul_group
```

## 장애인 지하철 역 이용량

```python
aug
```

```python
sub_avg = aug[["정류장", "위도", "경도", "노선"]]
sub_avg["평균이용량"] = round((aug["합계"]+sept["합계"]+oct["합계"])/3, 0).astype("int")
sub_avg
```

```python
seoul_subway =  folium.Map(location=[37.5665, 126.9780], zoom_start=12)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul_subway)

line_colors = {1:"#0032A0", 2:"#00B140", 3:"#FC4C02", 4:"#00A9EO", 5:"#A05EB5", 6:"#A9431E", 7:"#67823A", 8:"#E31C79", 9:"#8C8279"}

for index, station in sub_avg.iterrows():
  line_color = line_colors.get(station["노선"])

  folium.CircleMarker(
      location=[station["위도"], station["경도"]],
      popup=f'{station["정류장"]}- {station["평균이용량"]}',
      color=line_color, fill=True, fill_color=line_color, fill_opacity=0.8,
      radius = station["평균이용량"]*0.0007
  ).add_to(seoul_subway)

seoul_subway
```
