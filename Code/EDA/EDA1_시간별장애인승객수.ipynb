흠 계절성이 있나? aug -> sept -> oct 점점 증가하네  
8월에 날씨가 너무 더워서 힘들어서 그런가 암 튼

```python
aug_med = aug[(aug["합계"]>aug["합계"].median()-300) & (aug["합계"]<aug["합계"].median()+300)]
aug_med_time = aug[(aug["합계"]>aug["합계"].median()-300) & (aug["합계"]<aug["합계"].median()+300)].iloc[:,4:26]
```

```python
sept_med = sept[(sept["합계"]>sept["합계"].median()-300) & (sept["합계"]<sept["합계"].median()+300)]
sept_med_time = sept[(sept["합계"]>sept["합계"].median()-300) & (sept["합계"]<sept["합계"].median()+300)].iloc[:,4:26]
```

```python
oct_med = oct[(oct["합계"]>oct["합계"].median()-300) & (oct["합계"]<oct["합계"].median()+300)]
oct_med_time = oct[(oct["합계"]>oct["합계"].median()-300) & (oct["합계"]<oct["합계"].median()+300)].iloc[:,4:26]
```
