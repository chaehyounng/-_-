```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

## df1

```python
df1=df1.nlargest(20, 'Score')
route1 =  folium.Map(location=[37.5665, 126.9780], zoom_start=12)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(route1)

line_colors = {1:"#0032A0", 2:"#00B140", 3:"#FC4C02", 4:"#00A9EO", 5:"#A05EB5", 6:"#A9431E", 7:"#67823A", 8:"#E31C79", 9:"#8C8279"}

for index, station in df1.iterrows():
  line_color = line_colors.get(station["노선"])

  folium.CircleMarker(
      location=[station["위도"], station["경도"]],
      popup=f'{station["정류장"]}',
      color=line_color, fill=True, fill_color=line_color, fill_opacity=0.8,
      radius = station["Score"]*20
  ).add_to(route1)

route1
```
