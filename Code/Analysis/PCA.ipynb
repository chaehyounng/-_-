```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
oct_final=pd.read_csv("oct_final.csv")
oct_final
```

```python
oct_final_X=oct_final[['합계','주변_편의시설_개수','주변_장애인시설_개수','주거_장애인_수']]
oct_final_X
```

```python
# 주성분 분석 수행
pca = PCA()
pca.fit(oct_final_X)

# 주성분에 대한 설명력 확인
explained_variance_ratio = pca.explained_variance_ratio_

# 각 변수의 중요성을 기반으로 가중치 설정
weights = explained_variance_ratio / sum(explained_variance_ratio)

# 각 행에 대한 가중합 계산
oct_final_X['Score'] = oct_final_X.apply(lambda row: sum(row[col] * weights[i] for i, col in enumerate(oct_final_X.columns)), axis=1)

# 점수에 따라 내림차순 정렬 후 상위 몇 개 추출
top_rows = oct_final_X.sort_values(by='Score', ascending=False).head(3)

# 결과 출력
print(top_rows)
```
